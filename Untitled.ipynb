```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set font to display minus signs correctly (English only)
plt.rcParams['axes.unicode_minus'] = False

# Function to calculate Maximum Drawdown (MDD)
def calculate_mdd(returns):
    if returns.empty or returns.isna().all():
        return np.nan
    returns = returns.fillna(0)
    cum_returns = (1 + returns).cumprod()
    running_max = np.maximum.accumulate(cum_returns)
    drawdown = (cum_returns / running_max) - 1
    return drawdown.min()

# Function to calculate annualized volatility
def calculate_vol(returns, periods_per_year=252):
    if returns.empty or returns.isna().all():
        return np.nan
    returns = returns.fillna(0)
    return returns.std() * np.sqrt(periods_per_year)

# Generate sample return data
np.random.seed(42)
dates = pd.date_range('2020-01-01', periods=252)
returns_data = {
    'Strategy1': pd.Series(np.random.normal(0.0005, 0.01, len(dates)), index=dates),
    'Strategy2': pd.Series(np.random.normal(0.0007, 0.015, len(dates)), index=dates),
    'Strategy3': pd.Series(np.random.normal(0.0003, 0.008, len(dates)), index=dates)
}

# Calculate risk metrics
mdd = {s: calculate_mdd(r) for s, r in returns_data.items()}
vol = {s: calculate_vol(r) for s, r in returns_data.items()}

# Print results
print("=== Strategy Risk Metrics Comparison ===")
print("Maximum Drawdown (smaller = lower risk):")
print(pd.Series(mdd).sort_values().round(4))
print("\nAnnualized Volatility (smaller = more stable):")
print(pd.Series(vol).sort_values().round(4))

# Prepare data for plotting (convert to %)
df = pd.DataFrame({
    'Max Drawdown (%)': pd.Series(mdd) * 100,
    'Annualized Volatility (%)': pd.Series(vol) * 100
})

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# -------------------- Plot 1: Max Drawdown --------------------
colors1 = ['#ff6b6b', '#4ecdc4', '#45b7d1']
ax1.bar(df.index, df['Max Drawdown (%)'], color=colors1)
ax1.set_title('Maximum Drawdown Comparison (%)', fontsize=14)
ax1.set_ylabel('Max Drawdown (%)')
ax1.grid(axis='y', alpha=0.3)

# Add value labels on bars
for i, v in enumerate(df['Max Drawdown (%)']):
    ax1.text(i, v - 0.3, f'{v:.2f}%', ha='center', fontsize=10)

# -------------------- Plot 2: Annualized Volatility --------------------
colors2 = ['#ff9f43', '#10ac84', '#0abde3']
ax2.bar(df.index, df['Annualized Volatility (%)'], color=colors2)
ax2.set_title('Annualized Volatility Comparison (%)', fontsize=14)
ax2.set_ylabel('Annualized Volatility (%)')
ax2.grid(axis='y', alpha=0.3)

# Add value labels on bars
for i, v in enumerate(df['Annualized Volatility (%)']):
    ax2.text(i, v + 0.1, f'{v:.2f}%', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```
